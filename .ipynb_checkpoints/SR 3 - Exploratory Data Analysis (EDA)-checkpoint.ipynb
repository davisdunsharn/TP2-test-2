```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
infrastructure = pd.read_csv(r"C:\Users\Mlondi\OneDrive - Mangosuthu University of Technology\personal\personal\This pc\Documents\infrastructure_zaf.csv")
```

```python
# Basic infographic charts
# 1. Histogram of the 'Value' column for a specific range of rows (e.g., rows 50 to 55)
plt.figure(figsize=(10, 6))
plt.hist(infrastructure['Value'].loc[50:90], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Values for Rows 50 to 55')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
plt.show()
```

```python
#2. Boxplot of 'Value' by 'Year' using a subset of 10 rows
plt.figure(figsize=(12, 6))
subset_data = infrastructure.loc[0:9, ['Year', 'Value']]  # Selecting specific columns and 10 rows
sns.boxplot(x='Year', y='Value', data=subset_data)
plt.title('Boxplot')
```

```python
# 3. Bar chart of mean 'Value' by 'Indicator Name'

# Selecting a subset of the data (e.g., first 100 rows) before calculating the mean
subset_data = infrastructure.loc[300:500].copy()

# Convert 'Value' column to numeric
subset_data.loc[:, 'Value'] = pd.to_numeric(subset_data['Value'], errors='coerce')

# Bar chart of mean 'Value' by 'Indicator Name' for the subset
mean_values_subset = subset_data.groupby('Indicator Name')['Value'].mean().sort_values()
plt.figure(figsize=(10, 8))
mean_values_subset.plot(kind='barh', color='lightgreen')
plt.title('Mean Value by Indicator Name (Subset of Data)')
plt.xlabel('Mean Value')
plt.ylabel('Indicator Name')
plt.show()
```

```python
# Selecting a subset of the data (e.g., first 100 rows)
subset_data = infrastructure.head(100).copy()  # Make a copy to avoid SettingWithCopyWarning

# Convert 'Value' column to numeric
subset_data.loc[:, 'Value'] = pd.to_numeric(subset_data['Value'], errors='coerce')

# Example of an advanced chart: Heatmap to show correlation between different indicators
plt.figure(figsize=(12, 8))
indicator_pivot = subset_data.pivot_table(values='Value', index='Year', columns='Indicator Name')
sns.heatmap(indicator_pivot.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap between Indicators (Subset of Data)')
plt.show()
```

```python
# Selecting a subset of the data for demonstration (e.g., first 100 rows)
subset_data = infrastructure.tail(20).copy()  # Make a copy to avoid SettingWithCopyWarning

# Convert 'Value' column to numeric
subset_data.loc[:, 'Value'] = pd.to_numeric(subset_data['Value'], errors='coerce')

# Example of an advanced chart: Line plot with error bars to show variation over time
plt.figure(figsize=(12, 6))

# Group the data by 'Year' and calculate mean and standard deviation for each year
grouped_data = subset_data.groupby('Year')['Value'].agg(['mean', 'std'])

# Plotting the mean values with error bars representing the standard deviation
plt.errorbar(grouped_data.index, grouped_data['mean'], yerr=grouped_data['std'], fmt='-o', capsize=5)
plt.title('Mean Value with Standard Deviation Over Time')
plt.xlabel('Year')
plt.ylabel('Mean Value')
plt.grid(True)
plt.show()
```

```python
# Selecting a subset of the data for demonstration (e.g., first 100 rows)
subset_data = infrastructure.head(100).copy()  # Make a copy to avoid SettingWithCopyWarning

# Convert 'Value' column to numeric
subset_data.loc[:, 'Value'] = pd.to_numeric(subset_data['Value'], errors='coerce')

# Calculate summary statistics
mean_value = subset_data['Value'].mean()
median_value = subset_data['Value'].median()
std_deviation = subset_data['Value'].std()

# Print the summary statistics
print("Summary Statistics:")
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Standard Deviation: {std_deviation}")
```
